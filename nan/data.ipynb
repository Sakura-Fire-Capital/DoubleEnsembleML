```python
BTC=[]
for i in range(len(df)):
    row=df.iloc[i].values.tolist()
    BTC.append(row)
    print(row)
```

```python
for i in BTC:
    if i[1]<i[3]:
        i[0]=0
    else:
        i[0]=1
    del i[1]
    del i[3]
BTC
```

```python
len(BTC)
```

```python
testBTC=BTC[:265]
trainBTC=BTC[266:]
```

```python
with open('testBTC.txt', 'w') as f:
    for i in range(len(testBTC)):
        s = (re.sub(r"['{ },]*", '', str(testBTC[i])) + '\n').replace(':', ',')
        f.write(s)
    f.close()
```

```python
with open('trainBTC.txt', 'w') as m:
    for i in range(len(trainBTC)):
        s = (re.sub(r"['{ },]*", '', str(trainBTC[i])) + '\n').replace(':', ',')
        m.write(s)
    m.close()
```
